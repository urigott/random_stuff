# Is upsampling imbalanced data a good idea?

Here I'll try to show why up-sampling the minority class may alter the model's understanding of reality.

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.calibration import calibration_curve

from lightgbm import LGBMClassifier
```

To generate an imbalanced dataset, Let's start with creating a balanced dataset, but then removing many of the positive outcomes. We'll get back to them later...

```python
X, y = make_classification(n_samples=1_000_000, 
                           n_features=10, 
                           n_informative=10, 
                           n_redundant=0, 
                           random_state=42)

# add some noise to X
X = X * (1 + np.random.normal(size=X.shape))

X = pd.DataFrame(X)
y = pd.Series(y)

print(f'Original dataset is balanced, with outcome rate = {y.mean():.3f}\n\n')

# Remove positive cases from dataset, to create an imbalanced dataset.
np.random.seed(32)
mask = np.random.choice(y[y == 1].index, size=int(len(y) * 0.49), replace=False)
y_imbalanced = y.loc[~y.index.isin(mask)]
X_imbalanced = X.loc[~y.index.isin(mask)]

X_train, X_test, y_train, y_test = train_test_split(X_imbalanced, y_imbalanced, random_state=42)

print(f"""However, we remove many of the records with positive outcomes, so we end up with imbalanced dataset.
We then split it to train and test, with both datasets have similar outcome rates: {y_train.mean():.3f} and {y_test.mean():.3f}, respectively.
The test set should always represent reality. In our reality, the current dataset represents the imbalanced reality, in which the outcome rate is ~{y_test.mean():.2f}.
""")
```

### Modeling

We get pretty good results, but more importantly - our model is relatively calibrated, i.e., the outcome rate for records with 30% of outcome is 30%.
This is often extremely important, especially when absolute risk is necessary (for example, in the medical domain).


```python
clf = LGBMClassifier(max_depth=5, verbose=-1).fit(X_train, y_train)

train_pred = clf.predict_proba(X_train)[:, 1]
test_pred = clf.predict_proba(X_test)[:, 1]

train_auc = roc_auc_score(y_train, train_pred)
test_auc = roc_auc_score(y_test, test_pred)

fig, ax = plt.subplots(figsize=(6,6))
ax.grid()
ax.plot([0, 1], [0, 1], color='gray', alpha=0.6, ls='--')

obs, prd = calibration_curve(y_train, train_pred, n_bins=5)
ax.plot(obs, prd, label=f'Train (AUC: {train_auc:.3f})')

obs, prd = calibration_curve(y_test, test_pred, n_bins=5)
ax.plot(obs, prd, label=f'Test (AUC: {test_auc:.3f})')

ax.set_title(f'Imbalanced data | outcome rate in train set={y_train.mean():.3f}')
ax.set_xlabel('Predicted')
ax.set_ylabel('Observed')
ax.legend(loc=4);
```

### Upsampling

Let's upsample our training data and see what happens.
To do so, we'll add the positive samples that we previously removed into the train dataset.

**Note that we don't change the test set**, as it is a true representation of our reality, and on which we would like to test our model!

While we slightly (if any) improve our AUC results, the model becomes un-calibated for the test set. 




```python
X_upsampling = X.loc[y.index.isin(mask)]
y_upsampling = y.loc[y.index.isin(mask)]

X_train_new = pd.concat([X_train, X_upsampling])
y_train_new = pd.concat([y_train, y_upsampling])

clf = LGBMClassifier(max_depth=5, verbose=-1).fit(X_train_new, y_train_new)

train_pred = clf.predict_proba(X_train_new)[:, 1]
test_pred = clf.predict_proba(X_test)[:, 1]

train_auc = roc_auc_score(y_train_new, train_pred)
test_auc = roc_auc_score(y_test, test_pred)

fig, ax = plt.subplots(figsize=(6,6))
ax.grid()
ax.plot([0, 1], [0, 1], color='gray', alpha=0.6, ls='--')

obs, prd = calibration_curve(y_train_new, train_pred, n_bins=8)
ax.plot(obs, prd, label=f'Train (AUC: {train_auc:.3f})')

obs, prd = calibration_curve(y_test, test_pred, n_bins=8)
ax.plot(obs, prd, label=f'Test (AUC: {test_auc:.3f})')

ax.set_title(f'Balanced data | outcome rate in train set={y_train_new.mean():.3f}')
ax.set_xlabel('Predicted')
ax.set_ylabel('Observed')
ax.legend(loc=4);
```
